# 1. Exploratory Data Analysis (EDA)

This notebook performs comprehensive exploratory data analysis on the loan dataset to understand patterns, distributions, and relationships.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (14, 8)
```

## 1.1 Load Dataset

```python
# Load dataset
df = pd.read_csv('../data/loans.csv')

print(f"Dataset Shape: {df.shape}")
print(f"\nColumn Names & Types:")
print(df.dtypes)
print(f"\nFirst few rows:")
df.head()
```

## 1.2 Data Quality Check

```python
# Check for missing values
print("Missing Values:")
print(df.isnull().sum())

# Check for duplicates
print(f"\nDuplicate Rows: {df.duplicated().sum()}")

# Check data info
print(f"\nDataset Info:")
df.info()
```

## 1.3 Target Variable Distribution

```python
# Default distribution
default_counts = df['Default'].value_counts()
default_pct = df['Default'].value_counts(normalize=True) * 100

print("Default Distribution:")
print(f"No Default (0): {default_counts[0]:,} ({default_pct[0]:.2f}%)")
print(f"Default (1): {default_counts[1]:,} ({default_pct[1]:.2f}%)")

# Plot
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
df['Default'].value_counts().plot(kind='bar', color=['green', 'red'], alpha=0.7)
plt.title('Default Distribution', fontsize=12, fontweight='bold')
plt.xlabel('Default')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
df['Default'].value_counts(normalize=True).plot(kind='pie', autopct='%1.1f%%', 
                                                 colors=['green', 'red'], alpha=0.7)
plt.title('Default Percentage', fontsize=12, fontweight='bold')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

## 1.4 Numeric Features Distribution

```python
# Descriptive statistics
numeric_cols = df.select_dtypes(include=[np.number]).columns
print("Descriptive Statistics:")
df[numeric_cols].describe().round(2)
```

```python
# Distribution plots for numeric features
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for idx, col in enumerate(numeric_cols):
    if col != 'Default':
        axes[idx].hist(df[col], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
        axes[idx].set_title(f'{col} Distribution', fontsize=11, fontweight='bold')
        axes[idx].set_xlabel(col)
        axes[idx].set_ylabel('Frequency')
        axes[idx].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## 1.5 Categorical Features Distribution

```python
# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
print(f"Categorical columns: {list(categorical_cols)}")

# Plot distributions
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.flatten()

for idx, col in enumerate(categorical_cols):
    df[col].value_counts().plot(kind='bar', ax=axes[idx], color='steelblue', alpha=0.7)
    axes[idx].set_title(f'{col}', fontsize=11, fontweight='bold')
    axes[idx].set_ylabel('Count')
    axes[idx].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## 1.6 Correlation Analysis

```python
# Correlation with default
numeric_with_default = df.select_dtypes(include=[np.number]).columns
correlations = df[numeric_with_default].corr()['Default'].sort_values(ascending=False)

print("Correlation with Default:")
print(correlations)

# Plot correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df[numeric_with_default].corr(), annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1)
plt.title('Correlation Matrix', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 1.7 Key Insights Summary

```python
print("="*80)
print("KEY INSIGHTS FROM EDA")
print("="*80)

print("\n1. DATASET SIZE")
print(f"   Total Records: {len(df):,}")
print(f"   Features: {len(df.columns)}")
print(f"   Data Quality: 100% complete (no missing values)")

print("\n2. TARGET VARIABLE")
print(f"   Default Rate: {default_pct[1]:.2f}%")
print(f"   Class Imbalance: 7.6:1 (non-default:default)")

print("\n3. TOP PREDICTORS OF DEFAULT")
for i, (col, corr) in enumerate(correlations[1:6].items(), 1):
    direction = "↓" if corr < 0 else "↑"
    print(f"   {i}. {col}: {corr:.4f} {direction}")

print("\n4. NUMERIC FEATURES RANGE")
for col in numeric_cols:
    if col != 'Default':
        print(f"   {col}: [{df[col].min()}, {df[col].max()}]")

print("\n5. CATEGORICAL FEATURES")
for col in categorical_cols:
    print(f"   {col}: {df[col].nunique()} categories")
```
